```python
# Function to calculate the GCD of two numbers
def gcd(a, b):
    while b:
        a, b = b, a % b
    return a

# Input the two numbers
num1 = int(input("Enter the first number: "))
num2 = int(input("Enter the second number: "))

# Calculate the GCD using the gcd() function
result = gcd(num1, num2)

# Print the GCD
print("The greatest common divisor (GCD) of", num1, "and", num2, "is", result)
```

```python
# Function to find prime factors of a number
def prime_factors(n):
    factors = []
    divisor = 2
    while divisor <= n:
        if n % divisor == 0:
            factors.append(divisor)
            n //= divisor
        else:
            divisor += 1
    return factors

# Input the number
number = int(input("Enter a number: "))

# Find prime factors using list comprehension
factors = prime_factors(number)

# Print the prime factors
print("Prime Factors of", number, ":", factors)
```

```python
def sieve_eratosthenes(limit):
    sieve = [True] * (limit + 1)
    sieve[0] = sieve[1] = False
    for current in range(2, int(limit**0.5) + 1):
        if sieve[current]:
            for multiple in range(current * current, limit + 1, current):
                sieve[multiple] = False
    return [num for num, is_prime in enumerate(sieve) if is_prime]

# Input the limit
limit = int(input("Enter a limit: "))

# Generate a list of prime numbers using the Sieve of Eratosthenes algorithm
prime_numbers = sieve_eratosthenes(limit)

# Print the prime numbers
print("Prime Numbers up to", limit, ":", prime_numbers)
```

```python
# Define a function to calculate the factorial of a number
def factorial(n):
    if n == 0:
        return 1
    else:
        return n * factorial(n - 1)

# Calculate the sum of digits of factorial numbers from 1 to 5 using list comprehension
sum_of_digits = [sum(int(digit) for digit in str(factorial(num))) for num in range(1, 6)]

# Print the results
print("Factorial Numbers from 1 to 5:")
for num, sum_digits in enumerate(sum_of_digits, start=1):
    print(f"Factorial of {num}: {factorial(num)}, Sum of Digits: {sum_digits}")
```

```python
def is_prime(num):
    if num < 2:
        return False
    for i in range(2, int(num**0.5) + 1):
        if num % i == 0:
            return False
    return True

# Generate a list of prime palindromic numbers using list comprehension
prime_palindromic_numbers = [num for num in range(10, 1000) if is_prime(num) and str(num) == str(num)[::-1]]

# Print the prime palindromic numbers
print("Prime Palindromic Numbers from 10 to 999:", prime_palindromic_numbers)
```
