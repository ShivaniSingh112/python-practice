```python
#To count no of vowels
# taking string as input from user
str1 = input("enter the String : ")
# intialize a varible vowels tp zero
vowels = 0
#convert the string to lowercase so we have to put less condition in if statment
str1.lower()
#use a for loop to iterate over the string
for i in str1:
  #check for vowels
    if(i == 'a' or i == 'e' or i == 'i' or i == 'o' or i == 'u'):
      #increment varible vowel when condition become true
        vowels = vowels + 1
 #print the no of vowels
print("Total Number of Vowels in this String = ", vowels)
```

```python
#Reverse the string and print
# taking string as input from user
str2=input("enter the string:")
#use string slicing to reverse the string
st=str2[::-1]
#printing the reversed string
print("the reverse string is:",st)
```

```python
#To check a palindrome
# taking string as input from user
str3=input("enter the string")
#use string slicing to reverse the string
revstr=str3[::-1]
#check if both the strings are equal or not
if(str3==revstr):
  #print palindrome if strings are equal
  print("it is a palindrome")
else:
#print if not a palindrome
  print("it is not a palindrome")
```
